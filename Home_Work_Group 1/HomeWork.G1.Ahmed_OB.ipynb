### Calculate the frequency of each color

No, it is not possible to divide a categorical dataset into quartiles since quartiles are a measure of distribution for numerical data. Quartiles divide numerical data into four equal parts, each containing 25% of the data.

#### We can calculate frequencies or percentages for each category 

#### 1st Approach

```python
from collections import Counter

colors = ['Red', 'Green', 'Blue', 'Blue', 'Green', 'Red', 'Yellow', 'Orange', 'Purple', 'Blue', 'Green', 'Red', 'Blue', 'Yellow', 'Orange', 'Green', 'Blue', 'Red', 'Yellow', 'Green']

color_counts = Counter(colors)
print(color_counts)
```

#### 2nd Approach

```python
# Create a pandas DataFrame object from the Colors and Sizes lists
colors = pd.Series(['Red', 'Green', 'Blue', 'Blue', 'Green', 'Red', 'Yellow', 'Orange', 'Purple', 'Blue', 'Green', 'Red', 'Blue', 'Yellow', 'Orange', 'Green', 'Blue', 'Red', 'Yellow', 'Green'])
sizes = pd.Series(['Small', 'Medium', 'Large', 'Medium', 'Small', 'Small', 'Large', 'Medium', 'Medium', 'Large', 'Medium', 'Small', 'Large', 'Small', 'Large', 'Small', 'Medium', 'Medium', 'Small', 'Large'])
data = pd.DataFrame({'Colors': colors, 'Sizes': sizes})

# Use the crosstab() function to create a cross-tabulation table
cross_tab = pd.crosstab(data.Colors, data.Sizes)

# Print the cross-tabulation table
print(cross_tab)
```

#### 3rd Approach

```python
# Create a pandas Series object from the Colors list
colors = pd.Series(['Red', 'Green', 'Blue', 'Blue', 'Green', 'Red', 'Yellow', 'Orange', 'Purple', 'Blue', 'Green', 'Red', 'Blue', 'Yellow', 'Orange', 'Green', 'Blue', 'Red', 'Yellow', 'Green'])

# Calculate the mode of the dataset
mode = colors.mode()[0]

# Calculate the relative frequency of each category
relative_freq = colors.value_counts(normalize=True)

# Print the mode and relative frequency
print('Mode:', mode)
print('Relative frequency:\n', relative_freq)
```

#### 4th Approach

```python
# Create a pandas Series object from the Colors list
colors = pd.Series(['Red', 'Green', 'Blue', 'Blue', 'Green', 'Red', 'Yellow', 'Orange', 'Purple', 'Blue', 'Green', 'Red', 'Blue', 'Yellow', 'Orange', 'Green', 'Blue', 'Red', 'Yellow', 'Green'])

# Use the value_counts() method to count the frequency of each category
frequency_table = colors.value_counts()

# Create a bar chart using the plot() method
frequency_table.plot(kind='bar')

# Add axis labels and a title
plt.xlabel('Colors')
plt.ylabel('Frequency')
plt.title('Frequency of Colors')
plt.show()
```

#### 5th Approach

```python
# Create a pie chart using the plot() method
frequency_table.plot(kind='pie')

# Add a title
plt.title('Proportion of Colors')
plt.show()
```

Green and blue represent over half of the total colors in this pie chart.
