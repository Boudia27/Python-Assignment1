```python
# Generate some random data
data = np.random.normal(loc=10, scale=2, size=1000)

# Calculate mean, standard deviation, median, and mode
mean = np.mean(data)
std = np.std(data)
median = np.median(data)
mode = pd.Series(data).mode()[0]

# Calculate range, percentile 25, percentile 75, and interquartile range
range_ = np.max(data) - np.min(data)
prc25 = np.percentile(data, 25)
prc75 = np.percentile(data, 75)
iqr = prc75 - prc25

# Print the results
print(f"Mean: {mean:.2f}")
print(f"Standard deviation: {std:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode:.2f}")
print(f"Range: {range_:.2f}")
print(f"25th percentile: {prc25:.2f}")
print(f"75th percentile: {prc75:.2f}")
print(f"Interquartile range: {iqr:.2f}")
```
